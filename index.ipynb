```python
df=pd.read_csv('index (1).csv')
```

```python
print(" Data Loaded Successfully!")
print("Shape:", df.shape)
print("\nColumns:", df.columns.tolist())
print("\nFirst 5 rows:\n", df.head())
```

```python
print(df.info())
print(df.isnull().sum())
```

```python
essential_cols = ['date', 'datetime', 'cash_type', 'card', 'money', 'coffee_name']
df = df.dropna(subset=essential_cols)
```

```python
df['money'] = df['money'].fillna(0)
```

```python
df['coffee_name'] = df['coffee_name'].fillna('Unknown')
df['cash_type'] = df['cash_type'].fillna('Unknown')
df['card'] = df['card'].fillna('Unknown')
```

```python
print(df.isnull().sum())
```

```python
df['date'] = pd.to_datetime(df['date'], errors='coerce')
df['datetime'] = pd.to_datetime(df['datetime'], errors='coerce')
```

```python
df.dtypes
```

```python
df['hour']=df['datetime'].dt.hour
df['day']=df['datetime'].dt.day
df['month']=df['datetime'].dt.month
df['weekday']=df['datetime'].dt.day_name()
```

```python
df.describe()
```

```python
print("\nUnique Coffee Types:", df['coffee_name'].nunique())
print("Unique Cards:", df['card'].nunique())
```

```python
df['card'].value_counts()
```

```python
df['coffee_name'].value_counts()
```

```python
total_sales = df['money'].sum()
avg_sales = df['money'].mean()
print(f"\nTotal Sales: {total_sales:,.2f}")
print(f" Average Sale: {avg_sales:,.2f}")
```

```python
dir(sns)
```

```python
sns.histplot(df['money'],kde=True,bins=30,color='skyblue')
plt.title("Distribution Of Transaction Amounts")
plt.xlabel("Transaction Amount")
plt.ylabel("Count")
plt.show()
```

```python
coffee_sales = df.groupby('coffee_name')['money'].sum().sort_values(ascending=False)
sns.barplot(x=coffee_sales.values, y=coffee_sales.index, palette='viridis')
plt.title("Total Sales by Coffee Type")
plt.xlabel("Total Sales")
plt.ylabel("Coffee Type")
plt.show()
```

```python
hourly_sales = df.groupby('hour')['money'].sum()
sns.lineplot(x=hourly_sales.index, y=hourly_sales.values, marker='o', color='teal')
plt.title("Sales by Hour of Day")
plt.xlabel("Hour")
plt.ylabel("Total Sales")
plt.show()
```

```python
weekday_sales = df.groupby('weekday')['money'].sum()
order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
sns.barplot(x=weekday_sales.loc[order].index, y=weekday_sales.loc[order].values, palette='coolwarm')
plt.title("Sales by Day of Week")
plt.xlabel("Day of Week")
plt.ylabel("Total Sales")
plt.show()
```

```python
monthly_sales = df.groupby('month')['money'].sum()
sns.lineplot(x=monthly_sales.index, y=monthly_sales.values, marker='o', color='green')
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.show()
```

```python
top_customers = df.groupby('card')['money'].sum().sort_values(ascending=False).head(10)
sns.barplot(x=top_customers.values, y=top_customers.index, palette='magma')
plt.title("Top 10 Customers by Total Spend")
plt.xlabel("Total Spend")
plt.ylabel("Customer Card ID")
plt.show()
```

```python
sns.boxplot(x=df['money'], color='lightblue')
plt.title("Outlier Detection for Transaction Amounts")
plt.xlabel("Transaction Amount")
plt.show()
```

```python
num_cols = ['money', 'hour', 'day', 'month']
sns.heatmap(df[num_cols].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
hourly_sales = df.groupby('hour')['money'].sum()
sns.countplot(x='hour',data=df,hue="coffee_name")
plt.title("Sales by Hour of Day")
plt.xlabel("Hour")
plt.ylabel("Total Sales")
plt.show()
```
